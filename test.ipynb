```python
from game import Game2048, Action
import numpy as np

config = {
    'grid_size': 4,
    'tile_2_probability': 0.9,
    'initial_tiles': 2,
    'random_seed': 42,
}
game = Game2048(config)

step = 0
while not game.is_game_over():
    available = game.get_available_moves()
    if not available:
        break

    action = np.random.choice(available)
    valid, reward = game.move(action)
    step += 1

    print(f"--- Step {step} | Action: {action.name} | Reward: {reward} | Score: {game.get_score()} ---")
    print(game.get_state())
    print()

print(f"Game Over! Final Score: {game.get_score()} in {step} steps")
```
