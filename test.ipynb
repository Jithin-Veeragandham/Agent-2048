```python
from game import Game2048, Action
import numpy as np

config = {
    'grid_size': 4,
    'tile_2_probability': 0.9,
    'initial_tiles': 2,
    'random_seed': 42,
}

game = Game2048(config)
step = 0

while not game.is_game_over():
    available = game.get_available_moves()
    if not available:
        break

    # simulate each move on a clone, pick highest reward
    best_action = None
    best_reward = -1
    for action in available:
        sim = game.clone()
        _, reward = sim.move(action)
        if reward > best_reward:
            best_reward = reward
            best_action = action

    # tie-break: if all rewards are 0, pick randomly
    if best_reward == 0:
        best_action = available[np.random.randint(len(available))]

    valid, reward = game.move(best_action)
    step += 1
    print(f"--- Step {step} | Action: {best_action.name} | Reward: {reward} | Score: {game.get_score()} ---")
    print(game.get_state())
    print()

print(f"Game Over! Final Score: {game.get_score()} in {step} steps")
```
